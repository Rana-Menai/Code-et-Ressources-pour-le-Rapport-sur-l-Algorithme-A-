```python
import heapq

def a_star_search(graph, start, goal, heuristic):
    """
    Implémentation de l'algorithme A* pour la recherche de chemin optimal.

    Arguments:
    - graph: dict représentant le graphe (noeud -> voisins avec coûts)
    - start: noeud de départ
    - goal: noeud d'arrivée
    - heuristic: fonction heuristique h(n) estimant le coût restant

    Retourne:
    - Le chemin le plus court et son coût total
    """

    # File de priorité pour explorer les nœuds
    open_list = []
    heapq.heappush(open_list, (0, start))  # (coût total estimé, noeud)

    # Garde une trace des coûts
    g_costs = {start: 0}  # Coût réel pour atteindre chaque noeud
    came_from = {}  # Pour reconstruire le chemin

    while open_list:
        # Récupère le nœud avec le coût estimé le plus bas
        _, current = heapq.heappop(open_list)

        # Si le but est atteint, reconstruire le chemin
        if current == goal:
            path = []
            while current in came_from:
                path.append(current)
                current = came_from[current]
            path.append(start)
            return path[::-1], g_costs[goal]  # Chemin et coût total

        # Exploration des voisins
        for neighbor, cost in graph[current].items():
            tentative_g_cost = g_costs[current] + cost
            if neighbor not in g_costs or tentative_g_cost < g_costs[neighbor]:
                # Mise à jour du coût et du chemin
                g_costs[neighbor] = tentative_g_cost
                f_cost = tentative_g_cost + heuristic(neighbor, goal)
                heapq.heappush(open_list, (f_cost, neighbor))
                came_from[neighbor] = current

    return None, float('inf')  # Aucun chemin trouvé


# Définir le graphe comme un dictionnaire
graph = {
    'A': {'B': 1, 'C': 3},
    'B': {'A': 1, 'D': 2, 'E': 4},
    'C': {'A': 3, 'F': 5},
    'D': {'B': 2, 'E': 1},
    'E': {'B': 4, 'D': 1, 'F': 1},
    'F': {'C': 5, 'E': 1}
}

# Fonction heuristique (Manhattan distance simulée pour cet exemple)
def heuristic(node, goal):
    heuristics = {
        'A': 6, 'B': 4, 'C': 7,
        'D': 2, 'E': 1, 'F': 0  # 0 pour le noeud cible (goal)
    }
    return heuristics.get(node, float('inf'))

# Recherche du chemin le plus court de A à F
start_node = 'A'
goal_node = 'F'
path, cost = a_star_search(graph, start_node, goal_node, heuristic)

# Afficher les résultats
if path:
    print(f"Chemin trouvé: {' -> '.join(path)} avec un coût total de {cost}")
else:
    print("Aucun chemin trouvé.")
```

```python
import math
from heapq import heappush, heappop

class Noeud:
    def __init__(self, x, y, g=0, h=0):
        self.x = x
        self.y = y
        self.g = g  # Cout du chemin depuis le depart
        self.h = h  # Estimation heuristique vers l'objectif
        self.f = g + h  # Cout total
        self.parent = None  # Parent pour reconstituer le chemin

    def __lt__(self, other):
        return self.f < other.f

def distance_euclidienne(p1, p2):
    """Calcule la distance euclidienne entre deux points."""
    return math.sqrt((p1[0] - p2[0]) ** 2 + (p1[1] - p2[1]) ** 2)

def a_star(grille, depart, objectif):
    """Implementation de l'algorithme A*."""
    # Initialisation
    liste_ouverte = []  # File prioritaire (min-heap)
    liste_fermee = set()  # Ensemble des noeuds visites
    heappush(liste_ouverte, depart)

    while liste_ouverte:
        # a) Trouver le noeud avec le plus petit f
        q = heappop(liste_ouverte)

        # b) Ajouter q a la liste fermee
        liste_fermee.add((q.x, q.y))

        # c) Generer les successeurs de q
        directions = [
            (0, 1), (1, 0), (0, -1), (-1, 0),  # Directions cardinales
            (1, 1), (1, -1), (-1, 1), (-1, -1)  # Diagonales
        ]
        for dx, dy in directions:
            successeur_x = q.x + dx
            successeur_y = q.y + dy

            # Verifier les limites et l'accessibilite
            if not (0 <= successeur_x < len(grille) and 0 <= successeur_y < len(grille[0])):
                continue
            if grille[successeur_x][successeur_y] == 0:  # Case bloque
                continue

            successeur = Noeud(successeur_x, successeur_y)
            successeur.parent = q

            # d) Verifier si le successeur est l'objectif
            if (successeur.x, successeur.y) == (objectif.x, objectif.y):
                objectif.parent = q
                return reconstituer_chemin(objectif)

            # Calculer g, h et f pour le successeur
            successeur.g = q.g + distance_euclidienne((q.x, q.y), (successeur.x, successeur.y))
            successeur.h = distance_euclidienne((successeur.x, successeur_y), (objectif.x, objectif.y))
            successeur.f = successeur.g + successeur.h

            # iii) Ignorer le successeur s'il est dans la liste ouverte avec un cout inferieur
            if any(n for n in liste_ouverte if n.x == successeur.x and n.y == successeur.y and n.f <= successeur.f):
                continue

            # iv) Ignorer le successeur s'il est dans la liste fermee avec un cout inferieur
            if (successeur.x, successeur.y) in liste_fermee:
                continue

            # v) Ajouter le successeur a la liste ouverte
            heappush(liste_ouverte, successeur)

    raise Exception("Impossible de trouver un chemin vers l'objectif")

def reconstituer_chemin(noeud):
    """Reconstitue le chemin de l'objectif au depart."""
    chemin = []
    while noeud:
        chemin.append((noeud.x, noeud.y))
        noeud = noeud.parent
    return chemin[::-1]
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import math
from heapq import heappush, heappop

class Noeud:
    def __init__(self, nom, g=0, h=0):
        self.nom = nom
        self.g = g  # Coût du chemin depuis le départ
        self.h = h  # Heuristique estimée vers l'objectif
        self.f = g + h  # Coût total
        self.parent = None  # Parent pour reconstituer le chemin

    def __lt__(self, other):
        return self.f < other.f

def distance_euclidienne(p1, p2):
    """Calcule la distance euclidienne entre deux points."""
    return math.sqrt((p1[0] - p2[0]) ** 2 + (p1[1] - p2[1]) ** 2)

def a_star(graph, positions, depart, objectif):
    """Algorithme A* sur un graphe."""
    liste_ouverte = []
    liste_fermee = set()
    heappush(liste_ouverte, depart)

    etapes = []  # Pour visualiser les étapes
    while liste_ouverte:
        # Récupérer le noeud avec le plus petit f
        courant = heappop(liste_ouverte)
        liste_fermee.add(courant.nom)

        # Ajouter l'état actuel pour visualisation
        etapes.append({
            "courant": courant.nom,
            "g": courant.g,
            "h": courant.h,
            "f": courant.f,
            "liste_ouverte": [n.nom for n in liste_ouverte],
            "liste_fermee": list(liste_fermee)
        })

        # Vérifier si nous avons atteint l'objectif
        if courant.nom == objectif.nom:
            chemin = reconstituer_chemin(courant)
            return chemin, etapes

        # Explorer les voisins
        for voisin, poids in graph[courant.nom].items():
            if voisin in liste_fermee:
                continue

            successeur = Noeud(
                voisin,
                g=courant.g + poids["weight"],
                h=distance_euclidienne(positions[voisin], positions[objectif.nom]),
            )
            successeur.f = successeur.g + successeur.h
            successeur.parent = courant

            # Ajouter à la liste ouverte si nécessaire
            if all(n.nom != successeur.nom or n.f > successeur.f for n in liste_ouverte):
                heappush(liste_ouverte, successeur)

    return None, etapes

def reconstituer_chemin(noeud):
    """Reconstitue le chemin de l'objectif au départ."""
    chemin = []
    while noeud:
        chemin.append(noeud.nom)
        noeud = noeud.parent
    return chemin[::-1]

def afficher_graphe(graph, positions, chemin=None):
    """Affiche le graphe avec les positions."""
    plt.figure(figsize=(5, 3))
    nx.draw(graph, positions, with_labels=True, node_color='lightblue', font_weight='bold', node_size=700)
    if chemin:
        edges = [(chemin[i], chemin[i + 1]) for i in range(len(chemin) - 1)]
        nx.draw_networkx_edges(graph, positions, edgelist=edges, edge_color='red', width=2)
    plt.show()

def afficher_etapes(etapes):
    """Affiche les étapes sous forme de tableau."""
    print(f"{'Etape':<10}{'Courant':<10}{'g(n)':<10}{'h(n)':<10}{'f(n)':<10}{'Liste ouverte':<20}{'Liste fermee':<20}")
    for i, etape in enumerate(etapes):
        print(
            f"{i:<10}{etape['courant']:<10}{etape['g']:<10.2f}{etape['h']:<10.2f}{etape['f']:<10.2f}"
            f"{str(etape['liste_ouverte']):<20}{str(etape['liste_fermee']):<20}"
        )

# Exemple de graphe
graph = nx.Graph()
edges = [
    ("A", "B", 1),
    ("A", "C", 2),
    ("B", "D", 4),
    ("C", "D", 1),
    ("C", "E", 5),
    ("D", "F", 3),
    ("E", "F", 1),
]
graph.add_weighted_edges_from(edges)

# Positions pour le dessin du graphe
positions = {
    "A": (0, 0),
    "B": (1, 2),
    "C": (2, 0),
    "D": (3, 2),
    "E": (4, 0),
    "F": (5, 2),
}

# Définir les nœuds de départ et d'objectif
depart = Noeud("A", g=0, h=distance_euclidienne(positions["A"], positions["F"]))
objectif = Noeud("F")

# Exécution de l'algorithme A*
chemin, etapes = a_star(graph, positions, depart, objectif)

# Afficher le graphe et les résultats
afficher_graphe(graph, positions, chemin)
afficher_etapes(etapes)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import math
from heapq import heappush, heappop

class Noeud:
    def __init__(self, nom, g=0, h=0):
        self.nom = nom
        self.g = g  # Coût du chemin depuis le départ
        self.h = h  # Heuristique estimée vers l'objectif
        self.f = g + h  # Coût total
        self.parent = None  # Parent pour reconstituer le chemin

    def __lt__(self, other):
        return self.f < other.f

def distance_euclidienne(p1, p2):
    """Calcule la distance euclidienne entre deux points."""
    return math.sqrt((p1[0] - p2[0]) ** 2 + (p1[1] - p2[1]) ** 2)

def a_star(graph, positions, depart, objectif):
    """Algorithme A* sur un graphe."""
    liste_ouverte = []
    liste_fermee = set()
    heappush(liste_ouverte, depart)

    etapes = []  # Pour visualiser les étapes
    while liste_ouverte:
        # Récupérer le noeud avec le plus petit f
        courant = heappop(liste_ouverte)
        liste_fermee.add(courant.nom)

        # Ajouter l'état actuel pour visualisation
        etapes.append({
            "courant": courant.nom,
            "g": courant.g,
            "h": courant.h,
            "f": courant.f,
            "liste_ouverte": [n.nom for n in liste_ouverte],
            "liste_fermee": list(liste_fermee)
        })

        # Vérifier si nous avons atteint l'objectif
        if courant.nom == objectif.nom:
            chemin = reconstituer_chemin(courant)
            return chemin, etapes

        # Explorer les voisins
        for voisin, poids in graph[courant.nom].items():
            if voisin in liste_fermee:
                continue

            successeur = Noeud(
                voisin,
                g=courant.g + poids["weight"],
                h=distance_euclidienne(positions[voisin], positions[objectif.nom]),
            )
            successeur.f = successeur.g + successeur.h
            successeur.parent = courant

            # Ajouter à la liste ouverte si nécessaire
            if all(n.nom != successeur.nom or n.f > successeur.f for n in liste_ouverte):
                heappush(liste_ouverte, successeur)

    return None, etapes

def reconstituer_chemin(noeud):
    """Reconstitue le chemin de l'objectif au départ."""
    chemin = []
    while noeud:
        chemin.append(noeud.nom)
        noeud = noeud.parent
    return chemin[::-1]

def afficher_graphe(graph, positions, chemin=None):
    """Affiche le graphe avec les positions et les étiquettes des arêtes."""
    plt.figure(figsize=(5, 3))
    # Dessiner les nœuds et arêtes
    nx.draw(graph, positions, with_labels=True, node_color='lightblue', font_weight='bold', node_size=700)
    # Dessiner les étiquettes des arêtes
    edge_labels = nx.get_edge_attributes(graph, 'weight')
    nx.draw_networkx_edge_labels(graph, positions, edge_labels=edge_labels, font_color='green')
    if chemin:
        edges = [(chemin[i], chemin[i + 1]) for i in range(len(chemin) - 1)]
        nx.draw_networkx_edges(graph, positions, edgelist=edges, edge_color='red', width=2)
    plt.show()

def afficher_etapes(etapes):
    """Affiche les étapes sous forme de tableau."""
    print(f"{'Etape':<10}{'Courant':<10}{'g(n)':<10}{'h(n)':<10}{'f(n)':<10}{'Liste ouverte':<20}{'Liste fermee':<20}")
    for i, etape in enumerate(etapes):
        print(
            f"{i:<10}{etape['courant']:<10}{etape['g']:<10.2f}{etape['h']:<10.2f}{etape['f']:<10.2f}"
            f"{str(etape['liste_ouverte']):<20}{str(etape['liste_fermee']):<20}"
        )

# Exemple de graphe
graph = nx.Graph()
edges = [
    ("A", "B", 1),
    ("A", "C", 2),
    ("B", "D", 4),
    ("C", "D", 1),
    ("C", "E", 5),
    ("D", "F", 3),
    ("E", "F", 1),
]
graph.add_weighted_edges_from(edges)

# Positions pour le dessin du graphe
positions = {
    "A": (0, 0),
    "B": (1, 2),
    "C": (2, 0),
    "D": (3, 2),
    "E": (4, 0),
    "F": (5, 2),
}

# Définir les nœuds de départ et d'objectif
depart = Noeud("A", g=0, h=distance_euclidienne(positions["A"], positions["F"]))
objectif = Noeud("F")

# Exécution de l'algorithme A*
chemin, etapes = a_star(graph, positions, depart, objectif)

# Afficher le graphe et les résultats
afficher_graphe(graph, positions, chemin)
afficher_etapes(etapes)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import heapq
import pandas as pd

def astar_graphe_complexe(graphe, depart, objectif):
    def heuristique(n1, n2):
        # Heuristique simpliste basée sur la distance euclidienne (si disponible)
        x1, y1 = positions[n1]
        x2, y2 = positions[n2]
        return ((x1 - x2)**2 + (y1 - y2)**2)**0.5

    open_list = []
    closed_list = set()
    g_scores = {noeud: float('inf') for noeud in graphe.nodes}
    g_scores[depart] = 0
    heapq.heappush(open_list, (0 + heuristique(depart, objectif), depart))
    parent = {depart: None}

    exploration = []

    while open_list:
        _, courant = heapq.heappop(open_list)

        if courant in closed_list:
            continue

        closed_list.add(courant)

        if courant == objectif:
            chemin = []
            while courant:
                chemin.append(courant)
                courant = parent[courant]
            chemin.reverse()
            return chemin, exploration

        for voisin in graphe.neighbors(courant):
            tentative_g = g_scores[courant] + graphe[courant][voisin]['weight']
            if tentative_g < g_scores[voisin]:
                parent[voisin] = courant
                g_scores[voisin] = tentative_g
                f_score = tentative_g + heuristique(voisin, objectif)
                heapq.heappush(open_list, (f_score, voisin))
                exploration.append({
                    'Noeud': voisin,
                    'g(n)': tentative_g,
                    'h(n)': heuristique(voisin, objectif),
                    'f(n)': f_score,
                })

    return None, exploration  # Aucun chemin trouvé

# Création du graphe complexe
graphe = nx.Graph()
arcs = [
    ('A', 'B', 2), ('A', 'C', 4), ('B', 'D', 7), ('C', 'D', 1),
    ('C', 'E', 3), ('D', 'F', 5), ('E', 'F', 2), ('E', 'G', 8),
    ('F', 'G', 6)
]
positions = {
    'A': (0, 0), 'B': (1, 2), 'C': (2, 0), 'D': (3, 2),
    'E': (4, 0), 'F': (5, 2), 'G': (6, 0)
}
for arc in arcs:
    graphe.add_edge(arc[0], arc[1], weight=arc[2])

depart = 'A'
objectif = 'G'

# Exécution de l'algorithme A*
chemin, exploration = astar_graphe_complexe(graphe, depart, objectif)

# Visualisation du graphe et du chemin
plt.figure(figsize=(5, 3))
nx.draw(graphe, positions, with_labels=True, node_color='lightblue', node_size=1500, font_size=10)
labels = nx.get_edge_attributes(graphe, 'weight')
nx.draw_networkx_edge_labels(graphe, positions, edge_labels=labels)

if chemin:
    edges = [(chemin[i], chemin[i+1]) for i in range(len(chemin)-1)]
    nx.draw_networkx_edges(graphe, positions, edgelist=edges, edge_color='red', width=2)
    plt.title("Chemin trouvé par A*")
else:
    plt.title("Aucun chemin trouvé par A*")

plt.show()


# Résultats sous forme de tableau
df_exploration = pd.DataFrame(exploration)
print("Exploration des noeuds :")
print(df_exploration)

```

```python
import networkx as nx
import matplotlib.pyplot as plt
import math
from queue import PriorityQueue
import pandas as pd

def a_star_search(graph, start, goal):
    """
    Implémentation de l'algorithme A* sur un graphe donné.
    Retourne le chemin trouvé et un tableau récapitulatif des étapes d'exploration.
    """
    open_list = PriorityQueue()
    open_list.put((0, start))
    came_from = {}
    g_score = {node: float('inf') for node in graph.nodes}
    g_score[start] = 0
    f_score = {node: float('inf') for node in graph.nodes}
    f_score[start] = heuristic(start, goal)
    
    steps = []  # Stocke les étapes pour l'analyse

    while not open_list.empty():
        current = open_list.get()[1]
        
        # Sauvegarder les étapes
        steps.append({
            "current_node": current,
            "g_score": g_score.copy(),
            "f_score": f_score.copy(),
            "open_list": list(open_list.queue),
        })
        
        if current == goal:
            path = reconstruct_path(came_from, current)
            return path, steps
        
        for neighbor in graph.neighbors(current):
            tentative_g_score = g_score[current] + graph[current][neighbor]['weight']
            if tentative_g_score < g_score[neighbor]:
                came_from[neighbor] = current
                g_score[neighbor] = tentative_g_score
                f_score[neighbor] = g_score[neighbor] + heuristic(neighbor, goal)
                open_list.put((f_score[neighbor], neighbor))
    
    return None, steps

def heuristic(node, goal):
    """
    Fonction heuristique basée sur la distance euclidienne.
    """
    x1, y1 = positions[node]
    x2, y2 = positions[goal]
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

def reconstruct_path(came_from, current):
    """
    Reconstruit le chemin optimal à partir des parents.
    """
    path = [current]
    while current in came_from:
        current = came_from[current]
        path.append(current)
    path.reverse()
    return path

# Création du graphe complexe
graph = nx.Graph()

# Ajouter des nœuds
positions = {
    'A': (0, 0), 'B': (1, 3), 'C': (3, 1), 'D': (4, 4),
    'E': (5, 0), 'F': (6, 3), 'G': (7, 1)
}
graph.add_nodes_from(positions.keys())

# Ajouter des arêtes avec des poids
edges = [
    ('A', 'B', 4), ('A', 'C', 2), ('B', 'D', 5), ('C', 'D', 3),
    ('C', 'E', 6), ('D', 'F', 2), ('E', 'F', 1), ('E', 'G', 7),
    ('F', 'G', 4)
]
for edge in edges:
    graph.add_edge(edge[0], edge[1], weight=edge[2])

# Exécuter A* pour trouver le chemin
start_node, goal_node = 'A', 'G'
path, exploration_steps = a_star_search(graph, start_node, goal_node)

# Dessiner le graphe avec le chemin trouvé
plt.figure(figsize=(5, 3))
nx.draw(graph, positions, with_labels=True, node_color='lightblue', node_size=1500, font_size=10)
edge_labels = nx.get_edge_attributes(graph, 'weight')
nx.draw_networkx_edge_labels(graph, positions, edge_labels=edge_labels)

if path:
    path_edges = [(path[i], path[i + 1]) for i in range(len(path) - 1)]
    nx.draw_networkx_edges(graph, positions, edgelist=path_edges, edge_color='red', width=2)
    plt.title("Chemin trouvé par A* : " + " -> ".join(path))
else:
    plt.title("Aucun chemin trouvé par A*")

plt.show()

df_exploration = pd.DataFrame(exploration)
print("Exploration des noeuds :")
print(df_exploration)


'''# Générer le tableau des étapes d'exploration
table_data = []
for step in exploration_steps:
    current_node = step["current_node"]
    g_score = step["g_score"]
    f_score = step["f_score"]
    open_list = [item[1] for item in step["open_list"]]
    table_data.append({
        "Noeud actuel": current_node,
        "Scores g(n)": {k: round(v, 2) for k, v in g_score.items()},
        "Scores f(n)": {k: round(v, 2) for k, v in f_score.items()},
        "Liste ouverte": open_list
    })

# Afficher les étapes sous forme de tableau
for index, step in enumerate(table_data):
    print(f"\nÉtape {index + 1}:")
    print(f"Noeud actuel: {step['Noeud actuel']}")
    print("Scores g(n):", step["Scores g(n)"])
    print("Scores f(n):", step["Scores f(n)"])
    print("Liste ouverte:", step["Liste ouverte"])
'''
```

```python
import heapq
import math

class Noeud:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.g = float('inf')
        self.h = 0
        self.f = float('inf')
        self.parent = None

def distance_euclidienne(a, b):
    return math.sqrt((a.x - b.x)**2 + (a.y - b.y)**2)

def astar_navigation(graphe, depart, destination):
    open_list = []
    closed_list = set()
    heapq.heappush(open_list, (0, depart))

    depart.g = 0
    depart.h = distance_euclidienne(depart, destination)
    depart.f = depart.h

    while open_list:
        _, courant = heapq.heappop(open_list)
        if courant == destination:
            chemin = []
            while courant:
                chemin.append((courant.x, courant.y))
                courant = courant.parent
            return chemin[::-1]

        closed_list.add(courant)

        for voisin, cout in graphe.get(courant, []):
            if voisin in closed_list:
                continue
            tentative_g = courant.g + cout

            if tentative_g < voisin.g:
                voisin.g = tentative_g
                voisin.h = distance_euclidienne(voisin, destination)
                voisin.f = voisin.g + voisin.h
                voisin.parent = courant
                heapq.heappush(open_list, (voisin.f, voisin))

    return None

# Exemple de graphe
n1 = Noeud(0, 0)
n2 = Noeud(1, 1)
n3 = Noeud(2, 1)
n4 = Noeud(2, 3)
n5 = Noeud(3, 3)

graphe = {
    n1: [(n2, 1.4), (n3, 2)],
    n2: [(n4, 2.2)],
    n3: [(n4, 1)],
    n4: [(n5, 1)],
}

chemin = astar_navigation(graphe, n1, n5)
print("Chemin trouvé:", chemin)
```

```python
def creer_grille_obstacles(nb_lignes, nb_colonnes, obstacles):
    grille = [[1 for _ in range(nb_colonnes)] for _ in range(nb_lignes)]
    for (i, j) in obstacles:
        grille[i][j] = 0
    return grille

def astar_jeu(grille, source, destination):
    rows, cols = len(grille), len(grille[0])

    def heuristique(a, b):
        return abs(a[0] - b[0]) + abs(a[1] - b[1])  # Distance de Manhattan

    open_list = [(0, source)]
    closed_list = set()
    parents = {}
    g_scores = {source: 0}

    while open_list:
        _, courant = heapq.heappop(open_list)
        if courant == destination:
            chemin = []
            while courant in parents:
                chemin.append(courant)
                courant = parents[courant]
            chemin.append(source)
            return chemin[::-1]

        closed_list.add(courant)
        x, y = courant

        for dx, dy in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
            voisin = (x + dx, y + dy)
            if 0 <= voisin[0] < rows and 0 <= voisin[1] < cols and grille[voisin[0]][voisin[1]] == 1:
                if voisin in closed_list:
                    continue

                tentative_g = g_scores[courant] + 1

                if tentative_g < g_scores.get(voisin, float('inf')):
                    parents[voisin] = courant
                    g_scores[voisin] = tentative_g
                    f = tentative_g + heuristique(voisin, destination)
                    heapq.heappush(open_list, (f, voisin))

    return None

# Exemple
grille = creer_grille_obstacles(5, 5, [(1, 1), (2, 1), (2, 3)])
source = (0, 0)
destination = (4, 4)
chemin = astar_jeu(grille, source, destination)
print("Chemin trouvé:", chemin)
```

```python
import heapq

class EtatPuzzle:
    def __init__(self, grille, parent=None, g=0):
        self.grille = grille
        self.parent = parent
        self.g = g
        self.h = self.calculer_heuristique()
        self.f = self.g + self.h

    def __lt__(self, other):
        return self.f < other.f

    def calculer_heuristique(self):
        # Heuristique: Somme des distances de Manhattan
        h = 0
        for i in range(len(self.grille)):
            for j in range(len(self.grille[i])):
                if self.grille[i][j] != 0:
                    x, y = divmod(self.grille[i][j] - 1, 3)
                    h += abs(i - x) + abs(j - y)
        return h

    def generer_successeurs(self):
        successeurs = []
        x, y = [(i, j) for i, row in enumerate(self.grille) for j, val in enumerate(row) if val == 0][0]
        directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]

        for dx, dy in directions:
            nx, ny = x + dx, y + dy
            if 0 <= nx < 3 and 0 <= ny < 3:
                nouvelle_grille = [row[:] for row in self.grille]
                nouvelle_grille[x][y], nouvelle_grille[nx][ny] = nouvelle_grille[nx][ny], nouvelle_grille[x][y]
                successeurs.append(EtatPuzzle(nouvelle_grille, parent=self, g=self.g + 1))
        return successeurs

def astar_puzzle(puzzle_depart, puzzle_cible):
    def est_egal(grille1, grille2):
        return grille1 == grille2

    etat_initial = EtatPuzzle(puzzle_depart)
    cible = EtatPuzzle(puzzle_cible)

    open_list = []
    closed_list = set()
    heapq.heappush(open_list, etat_initial)

    while open_list:
        courant = heapq.heappop(open_list)

        if est_egal(courant.grille, cible.grille):
            chemin = []
            while courant:
                chemin.append(courant.grille)
                courant = courant.parent
            return chemin[::-1]

        closed_list.add(tuple(map(tuple, courant.grille)))

        for successeur in courant.generer_successeurs():
            if tuple(map(tuple, successeur.grille)) in closed_list:
                continue
            heapq.heappush(open_list, successeur)

    return None

# Exemple
puzzle_depart = [[1, 2, 3], [4, 0, 5], [6, 7, 8]]
puzzle_cible = [[1, 2, 3], [4, 5, 0], [6, 7, 8]]
chemin = astar_puzzle(puzzle_depart, puzzle_cible)
for etape in chemin:
    print(etape)
    print("---")
```

```python
from pyamaze import maze, agent, COLOR, textLabel
from queue import PriorityQueue

# Définition de la fonction heuristique
def h(cell1, cell2):
    x1, y1 = cell1
    x2, y2 = cell2
    return abs(x1 - x2) + abs(y1 - y2)

# Définition de la fonction A* pour la recherche de chemin
def aStar(m, start=None):
    # Si le point de départ n’est pas spécifié, le définir comme la taille du labyrinthe
    if start is None:
        start = (m.rows, m.cols)

    # Initialisation de la file de priorité
    open = PriorityQueue()
    open.put((h(start, m._goal), h(start, m._goal), start))

    # Initialisation des structures de données
    aPath = {}
    g_score = {row: float("inf") for row in m.grid}
    g_score[start] = 0
    f_score = {row: float("inf") for row in m.grid}
    f_score[start] = h(start, m._goal)

    searchPath = []

    # Boucle principale de l’algorithme A*
    while not open.empty():
        currCell = open.get()[2]
        searchPath.append(currCell)

        # Vérifier si la cellule actuelle est la destination
        if currCell == m._goal:
            break

        # Explorer les voisins dans les directions ’ESNW’
        for d in 'ESNW':
            if m.maze_map[currCell][d] == True:
                if d == 'E':
                    childCell = (currCell[0], currCell[1] + 1)
                elif d == 'W':
                    childCell = (currCell[0], currCell[1] - 1)
                elif d == 'N':
                    childCell = (currCell[0] - 1, currCell[1])
                elif d == 'S':
                    childCell = (currCell[0] + 1, currCell[1])

                # Calcul des scores temporaires
                temp_g_score = g_score[currCell] + 1
                temp_f_score = temp_g_score + h(childCell, m._goal)

                # Mise à jour des scores si un chemin plus court est trouvé
                if temp_f_score < f_score[childCell]:
                    aPath[childCell] = currCell
                    g_score[childCell] = temp_g_score
                    f_score[childCell] = temp_f_score
                    open.put((f_score[childCell], h(childCell, m._goal), childCell))

    # Reconstruction des chemins
    fwdPath = {}
    cell = m._goal
    while cell != start:
        fwdPath[aPath[cell]] = cell
        cell = aPath[cell]

    # Retour des chemins
    return searchPath, aPath, fwdPath
```

```python
import random
from pyamaze import maze, agent, COLOR
from queue import PriorityQueue

# Définir la fonction heuristique pour A*
def h(cell1, cell2):
    x1, y1 = cell1
    x2, y2 = cell2
    return abs(x1 - x2) + abs(y1 - y2)

# Implémentation de l'algorithme A*
def aStar(m, start=None):
    if start is None:
        start = (m.rows, m.cols)

    open = PriorityQueue()
    open.put((h(start, m._goal), h(start, m._goal), start))
    
    aPath = {}
    g_score = {cell: float("inf") for cell in m.grid}
    g_score[start] = 0
    f_score = {cell: float("inf") for cell in m.grid}
    f_score[start] = h(start, m._goal)
    searchPath = [start]
    
    while not open.empty():
        currCell = open.get()[2]
        searchPath.append(currCell)
        
        if currCell == m._goal:
            break
        
        for d in 'ESNW':
            if m.maze_map[currCell][d]:
                if d == 'E':
                    childCell = (currCell[0], currCell[1] + 1)
                elif d == 'W':
                    childCell = (currCell[0], currCell[1] - 1)
                elif d == 'N':
                    childCell = (currCell[0] - 1, currCell[1])
                elif d == 'S':
                    childCell = (currCell[0] + 1, currCell[1])
                
                temp_g_score = g_score[currCell] + 1
                temp_f_score = temp_g_score + h(childCell, m._goal)
                
                if temp_f_score < f_score[childCell]:
                    aPath[childCell] = currCell
                    g_score[childCell] = temp_g_score
                    f_score[childCell] = temp_f_score
                    open.put((f_score[childCell], h(childCell, m._goal), childCell))
    
    fwdPath = {}
    cell = m._goal
    while cell != start:
        fwdPath[aPath[cell]] = cell
        cell = aPath[cell]
    
    return searchPath, aPath, fwdPath

# Jeu principal
if __name__ == "__main__":
    # Création du labyrinthe
    m = maze(20, 20)
    m.CreateMaze(loopPercent=20)

    # Définir un but aléatoire
    goal_row = random.randint(1, m.rows)
    goal_col = random.randint(1, m.cols)
    m._goal = (goal_row, goal_col)

    # Résolution avec A*
    searchPath, aPath, fwdPath = aStar(m)

    # Agents et couleurs
    # Point de départ
    start_agent = agent(m, footprints=False, shape='square', color=COLOR.yellow)  # Départ jaune clair

    # Chemin exploré en points rouges
    search_agent = agent(m, footprints=True, color=COLOR.red)  # Tracer chaque cellule visitée
    m.tracePath({search_agent: searchPath}, showMarked=False)

    # But (agrandi et visible)
    goal_agent = agent(m, footprints=False, shape='square', color=COLOR.blue, goal=m._goal)  # But bleu large

    # Tracer le chemin optimal trouvé (vert clair)
    optimal_agent = agent(m, footprints=True, color=COLOR.green)
    m.tracePath({optimal_agent: fwdPath}, showMarked=True)

    # Afficher le labyrinthe
    m.run()
```

```python
import heapq

def a_star_with_metrics(graph, start, goal):
    def heuristic(node, goal):
        return abs(goal - node)  # Exemple simple (distance absolue)

    open_list = []
    heapq.heappush(open_list, (0, start))
    came_from = {}
    g_score = {node: float('inf') for node in graph}
    g_score[start] = 0
    f_score = {node: float('inf') for node in graph}
    f_score[start] = heuristic(start, goal)

    explored_nodes = 0
    total_cost = 0

    while open_list:
        current = heapq.heappop(open_list)[1]
        explored_nodes += 1

        if current == goal:
            path = []
            while current in came_from:
                path.append(current)
                current = came_from[current]
            path.append(start)
            path.reverse()
            total_cost = sum(graph[path[i]][path[i + 1]] for i in range(len(path) - 1))
            return {
                "path": path,
                "explored_nodes": explored_nodes,
                "total_cost": total_cost,
                "path_length": len(path),
            }

        for neighbor in graph[current]:
            tentative_g_score = g_score[current] + graph[current][neighbor]
            if tentative_g_score < g_score[neighbor]:
                came_from[neighbor] = current
                g_score[neighbor] = tentative_g_score
                f_score[neighbor] = tentative_g_score + heuristic(neighbor, goal)
                heapq.heappush(open_list, (f_score[neighbor], neighbor))

    return {
        "path": None,
        "explored_nodes": explored_nodes,
        "total_cost": float('inf'),
        "path_length": 0,
    }


# Exemple de graphe (représentation par dictionnaire)
graph = {
    0: {1: 1, 2: 4},
    1: {0: 1, 3: 2, 4: 7},
    2: {0: 4, 4: 3},
    3: {1: 2, 5: 1},
    4: {1: 7, 2: 3, 5: 2},
    5: {3: 1, 4: 2}
}

start = 0
goal = 5

# Exécution de l'algorithme
result = a_star_with_metrics(graph, start, goal)

# Affichage des métriques
print("Chemin trouvé :", result["path"])
print("Nombre de nœuds explorés :", result["explored_nodes"])
print("Coût total du chemin :", result["total_cost"])
print("Longueur du chemin :", result["path_length"])
```

```python
import heapq
import numpy as np
import matplotlib.pyplot as plt
import random

def a_star_dynamic(start, goal, grid):
    def heuristic(a, b):
        return abs(a[0] - b[0]) + abs(a[1] - b[1])
    
    def get_neighbors(node):
        neighbors = []
        for dx, dy in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
            nx, ny = node[0] + dx, node[1] + dy
            if 0 <= nx < len(grid) and 0 <= ny < len(grid[0]) and grid[nx][ny] != 1:
                neighbors.append((nx, ny))
        return neighbors

    open_list = []
    heapq.heappush(open_list, (0, start))
    came_from = {}
    g_score = {start: 0}
    
    while open_list:
        current = heapq.heappop(open_list)[1]
        if current == goal:
            path = []
            while current in came_from:
                path.append(current)
                current = came_from[current]
            path.append(start)
            return path[::-1]

        for neighbor in get_neighbors(current):
            tentative_g_score = g_score[current] + 1
            if neighbor not in g_score or tentative_g_score < g_score[neighbor]:
                came_from[neighbor] = current
                g_score[neighbor] = tentative_g_score
                f_score = tentative_g_score + heuristic(neighbor, goal)
                heapq.heappush(open_list, (f_score, neighbor))
    return None

# Create a dynamic grid (1 = obstacle, 0 = free)
grid_size = 20
grid = np.zeros((grid_size, grid_size), dtype=int)

# Add random obstacles
for _ in range(100):
    x, y = random.randint(0, grid_size - 1), random.randint(0, grid_size - 1)
    grid[x, y] = 1

start = (0, 0)
goal = (grid_size - 1, grid_size - 1)

# Apply A* and plot
path = a_star_dynamic(start, goal, grid)

plt.figure(figsize=(8, 5))
plt.imshow(grid, cmap='Greys', origin='lower')
if path:
    for x, y in path:
        plt.scatter(y, x, c='red')
plt.title("Dynamic Maze Navigation with A*")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def a_star_graph(graph, start, goal):
    def heuristic(node, goal):
        return nx.shortest_path_length(graph, node, goal, weight='weight')

    open_list = [(0, start)]
    came_from = {}
    g_score = {node: float('inf') for node in graph.nodes}
    g_score[start] = 0
    
    while open_list:
        _, current = heapq.heappop(open_list)
        if current == goal:
            path = []
            while current in came_from:
                path.append(current)
                current = came_from[current]
            path.append(start)
            return path[::-1]

        for neighbor in graph.neighbors(current):
            tentative_g_score = g_score[current] + graph[current][neighbor]['weight']
            if tentative_g_score < g_score[neighbor]:
                came_from[neighbor] = current
                g_score[neighbor] = tentative_g_score
                f_score = tentative_g_score + heuristic(neighbor, goal)
                heapq.heappush(open_list, (f_score, neighbor))
    return None

# Create a weighted graph with dynamic weights
G = nx.grid_2d_graph(10, 10)
for (u, v) in G.edges:
    G.edges[u, v]['weight'] = random.randint(1, 10)

start = (0, 0)
goal = (9, 9)

path = a_star_graph(G, start, goal)

# Visualize the graph
pos = {node: (node[1], -node[0]) for node in G.nodes}
weights = [G[u][v]['weight'] for u, v in G.edges]

plt.figure(figsize=(8, 5))
nx.draw(G, pos, with_labels=True, node_size=700, font_size=8, edge_color=weights, edge_cmap=plt.cm.Blues)
if path:
    nx.draw_networkx_edges(G, pos, edgelist=list(zip(path, path[1:])), edge_color='red', width=2)
plt.title("Traffic Simulation on Weighted Graph")
plt.show()
```
